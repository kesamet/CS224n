## 3 GRU

### Modelling latching behaviour

#### RNN
- Case 1: $x^{(t)}=0, h^{(t-1)}=0 \Rightarrow h^{(t-1)}=0$
$$ \sigma(b_h) = 0 \Rightarrow b_h \le 0 $$

- Case 2: $x^{(t)}=0, h^{(t-1)}=1 \Rightarrow h^{(t-1)}=1$
$$ \sigma(U_h+b_h) = 1 \Rightarrow U_h+b_h > 0 $$

- Case 3: $x^{(t)}=1, h^{(t-1)}=0 \Rightarrow h^{(t-1)}=1$
$$ \sigma(W_h+b_h) = 1 \Rightarrow W_h+b_h > 0 $$

- Case 4: $x^{(t)}=1, h^{(t-1)}=1 \Rightarrow h^{(t-1)}=1$
$$ \sigma(W_h+U_h+b_h) = 1 \Rightarrow W_h+U_h+b_h > 0 $$

Thus,
$$ U_h+b_h > 0, \quad W_h+b_h > 0, \quad b_h \le 0 $$

#### GRU
Since $W_r=U_r=b_r=b_z=b_h=0$,
$$ z^{(t)} = \sigma(x^{(t)}W_z + h^{(t-1)}U_z) $$
$$ r^{(t)} = 0 $$
$$ \tilde{h}^{(t)} = \tanh(x^{(t)}W_h) $$
$$ h^{(t)} = z^{(t)}h^{(t-1)} + (1-z^{(t)})\tanh(x^{(t)}W_h) $$

- Case 1: $x^{(t)}=0, h^{(t-1)}=0 \Rightarrow h^{(t-1)}=0$
$$ U_z, U_h, W_z, W_h \in \mathbb{R} $$

- Case 2: $x^{(t)}=0, h^{(t-1)}=1 \Rightarrow h^{(t-1)}=1$
$$ \sigma(U_z) = 1 \Rightarrow U_z > 0 $$

- Case 3: $x^{(t)}=1, h^{(t-1)}=0 \Rightarrow h^{(t-1)}=1$
$$ (1-\sigma(W_z))\tanh(x^{(t)}W_h) = 1 \Rightarrow W_h > 0, W_z \le 0 $$

- Case 4: $x^{(t)}=1, h^{(t-1)}=1 \Rightarrow h^{(t-1)}=1$
$$  U_h \in \mathbb{R} $$

Thus,
$$ U_z > 0, \quad U_h \in \mathbb{R}, \quad W_h > 0, \quad W_z \le 0 $$

### Modelling toggling behaviour

#### RNN
- Case 1: $x^{(t)}=0, h^{(t-1)}=0 \Rightarrow h^{(t-1)}=0$
$$ \sigma(b_h) = 0 \Rightarrow b_h \le 0 $$

- Case 2: $x^{(t)}=0, h^{(t-1)}=1 \Rightarrow h^{(t-1)}=1$
$$ \sigma(U_h+b_h) = 1 \Rightarrow U_h+b_h > 0 $$

- Case 3: $x^{(t)}=1, h^{(t-1)}=0 \Rightarrow h^{(t-1)}=1$
$$ \sigma(W_h+b_h) = 1 \Rightarrow W_h+b_h > 0 $$

- Case 4: $x^{(t)}=1, h^{(t-1)}=1 \Rightarrow h^{(t-1)}=0$
$$ \sigma(W_h+U_h+b_h) = 0 \Rightarrow W_h+U_h+b_h \le 0 $$

We have a contradiction.

#### GRU
Since $W_r=U_r=b_z=b_h=0$,
$$ z^{(t)} = \sigma(x^{(t)}W_z + h^{(t-1)}U_z) $$
$$ r^{(t)} = \sigma(b_r) $$
$$ \tilde{h}^{(t)} = \tanh(x^{(t)}W_h + r^{(t)}h^{(t-1)}U_h) $$
$$ h^{(t)} = z^{(t)}h^{(t-1)} + (1-z^{(t)})\tanh(x^{(t)}W_h + r^{(t)}h^{(t-1)}U_h) $$

- Case 1: $x^{(t)}=0, h^{(t-1)}=0 \Rightarrow h^{(t-1)}=0$
$$ U_z, U_h, W_z, W_h \in \mathbb{R} $$

- Case 2: $x^{(t)}=0, h^{(t-1)}=1 \Rightarrow h^{(t-1)}=1$
$$ \sigma(U_z) + (1-\sigma(U_z))\tanh(\sigma(b_r)U_h) = 1 $$

- Case 3: $x^{(t)}=1, h^{(t-1)}=0 \Rightarrow h^{(t-1)}=1$
$$ \sigma(W_z) + (1-\sigma(W_z))\tanh(W_h) = 1 $$

- Case 4: $x^{(t)}=1, h^{(t-1)}=1 \Rightarrow h^{(t-1)}=0$
$$ \sigma(W_z+U_z) + (1-\sigma(W_z+U_z))\tanh(W_h + \sigma(b_r)U_h) = 0 $$
